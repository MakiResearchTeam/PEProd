```python
from petools.cpu import PosePredictorLite
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
pose_predictor = PosePredictorLite(
    path_to_tflite='model/model.tflite',
    path_to_config='model/config.json',
    norm_mode='caffe'
)
```

## Try to predict

```python
img_path = '/mnt/data/labeling_service/left_del_2020-10-03 11-47-00/images/1126.png'
```

```python
single_img = cv2.imread(img_path)
```

```python
plt.imshow(single_img)
```

```python
predictions = pose_predictor.predict(single_img)
```

```python
predictions
```

```python
plt.imshow(pose_predictor.draw(single_img, predictions))
```

## Measure speed

```python
import time

start = time.time()

n = 100
for _ in range(n):
    _ = pose_predictor.predict(single_img)
print('fps: ', n / (time.time() - start))
```
